**3. Tidy Data**

3.1.1 Identify the variables in the dataset.

- country: Name of the Country

- iso2 & iso3: Country code.

- year: Year of the data was recorded.

- Other columns are also variables



  3.1.2 Perform a pivot operation to make the data tidy, storing the result in who2.

  3.1.3 Separate values like new_ep_f014 into components (e.g., new, ep, f014). Remove the column containing new, and store the result in who3.

  3.1.4 Further separate values like f014 into f and 014, storing the result in who_tidy.

```r
# Pivot operation
who2 <- who %>%
  pivot_longer(
    cols = new_sp_m014:newrel_f65,
    names_to = "key",
    values_to = "cases",
    values_drop_na = TRUE
  )

# Separate and remove 'new'
# Fix the 'newrel' inconsistency first
who3 <- who2 %>%
  mutate(key = str_replace(key, "newrel", "new_rel")) %>%
  separate(key, c("new", "type", "sexage"), sep = "_") %>%
  select(-new, -iso2, -iso3)

# 3.1.4 Separate sex and age
who_tidy <- who3 %>%
  separate(sexage, c("sex", "age"), sep = 1)
```

**3.2 Population Dataset**

  3.2.1 Identify the variables in this dataset.
  - country: Country name
  - year: Year of the data recorded
  - population: Population of  the Country

  3.2.2 Perform a pivot operation to tidy the data, storing the result in population3.

  3.2.3 Cast the population variable to an appropriate data type, storing the result in population_tidy.

```r
# Section 3.2: Clean the Population data
# 3.2.1 & 3.2.2 Read and Pivot
population3 <- population %>%
  pivot_longer(
    cols = -c("Country Name", "Country Code", "Indicator Name", "Indicator Code"),
    names_to = "year",
    values_to = "population"
  )

head(population3)
```

```r
# Fix data types and filter out the extra column
population_tidy <- population3 %>%
  # 1. Filter out rows that don't have a proper year (like ...69 or Unnamed)
  # str_detect looks for rows that start with "..."
  filter(!str_detect(year, "\\.\\.\\.")) %>%

  # 2. Now perform the conversion
  mutate(
    year = as.integer(readr::parse_number(year)),
    population = as.numeric(population)
  ) %>%
  rename(country="Country Name")

# Check the summary again - the warning should be gone
summary(population_tidy)
```

```r
# Find the difference of country name between the two dataset
setdiff(unique(who_tidy$country), unique(population_tidy$country))
```

**4. Data Manipulation**




  4.1 Determine the total TB cases among men and women in the 21st century in the United States. Identify which sex had more cases.



```r
us_21st <- tuberculosis %>%
  filter(country == "United States", year >= 2000) %>%
  group_by(sex) %>%
  summarise(total_cases = sum(cases))

us_21st

more_cases_sex <- us_21st$sex[which.max(us_21st$total_cases)]

cat("The sex that had more TB cases is:", ifelse(more_cases_sex == "m", "Men", "Women"))
```

  4.2 Create a new variable, cases_per_100k, representing TB cases per 100,000 people by year, sex, age group, and TB type.

  4.3 Identify:
- The country and year with the highest cases per 100k.
- The country and year with the lowest cases per 100k.

```r
tuberculosis <- tuberculosis %>%
  mutate(cases_per_100k = (cases / population) * 100000)

# Highest and Lowest
highest_rate <- tuberculosis %>% slice_max(cases_per_100k, n = 1) %>%
  slice(1) %>%
  select(country, sex, year, cases_per_100k)

lowest_rate <- tuberculosis %>% slice_min(cases_per_100k, n = 1) %>%
  slice(1) %>%
  select(country, sex, year, cases_per_100k)

cat("\n\033[1mHighest TB Rate\033[0m\n")
highest_rate

cat("\n\n\n\033[1mLowest TB Rate\033[0m\n")
lowest_rate
```

**5. Data Visualization**

5.1 Plot the total cases per 100k as a function of year for China, India, and the United States:
- Use a log scale on the y-axis (scale_y_log10()).
- Describe emerging patterns.


```r
tuberculosis %>%
  filter(country %in% c("China", "India", "United States")) %>%
  group_by(country, year, population) %>%
  summarise(total_rate = sum(cases_per_100k)) %>%
  ggplot(aes(x = year, y = total_rate, color = country)) +
  geom_line() +
  scale_y_log10() +
  labs(title = "TB Rates Over Time", y = "Cases per 100k (Log Scale)")+
  theme_minimal()
```


5.2 Compare distributions of total cases per 100k (summed over years, sexes, and TB types) across age groups:

- Use a log scale on the y-axis.
- Highlight observed patterns.



```r
ggplot(tuberculosis, aes(x = age, y = cases_per_100k)) +
  geom_boxplot() +
  scale_y_log10() +
  labs(title = "TB Rates vs Population Size (Year 2000)",
       x = "Age",
       y = "Cases per 100k (Log Scale)") +
  theme_minimal()
```

5.3 Create a plot to evaluate whether the number of cases per 100k in 2000 was related to a country’s population:


Conclude based on the visualization.


```r
tuberculosis %>%
  filter(year == 2000) %>%
  group_by(country, population) %>%
  summarise(total_rate = sum(cases_per_100k)) %>%
  ggplot(aes(x = population, y = total_rate)) +
  geom_point(alpha = 0.5) +
  scale_x_log10() +
  scale_y_log10() +
  geom_smooth(method = "lm", se = FALSE) +
  labs(title = "TB Rates vs Population Size",
       x = "Population (Log Scale)",
       y = "Cases per 100k (Log Scale)") +
  theme_minimal()
```

There isn't a strong linear correlation between a country's total population size and its TB rate, as the rate is more heavily influenced by healthcare infrastructure and socioeconomic factors. Big countries can also have high or low TB rate.
