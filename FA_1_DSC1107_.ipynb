# **Lecture 2 - Exercise 7.1 Relating the carat and cut of a diamond**

```r
library(tidyverse)
diamonds
```

## What relationship does it suggest between carat and cut? Create a plot to directly visualize this relationship.

- The plot suggests an inverse relationship between the quality of the cut and the size (carat) of the diamond. On average, lower-quality cuts (Fair) have significantly higher carat weights than higher-quality cuts (Ideal).


```r
# Visualizing the relationship between carat and cut
ggplot(data = diamonds, mapping = aes(x = cut, y = carat)) +
  geom_boxplot() +
  labs(title = "Distribution of Carat Weight by Cut Quality",
       subtitle = "Lower quality cuts tend to be larger (heavier) diamonds",
       x = "Cut Quality",
       y = "Carat (Weight)")
```

## What do you conclude? How does this explain the paradoxical trend we found in the plot below?

- Based on the graph below, where lower-quality cuts appear to have higher median prices, this is explained by the presence of carat weight as a confounding variable. While cut quality does influence price, carat weight is a far more dominant predictor of a diamond's value. The boxplot visualization reveals that "Fair" and "Good" diamonds in this dataset are, on average, significantly larger than "Ideal" diamonds. Because these lower-quality stones are much heavier, their total market price remains high despite the "quality discount" applied to their cut. Essentially, the price plot is not comparing like with like; it is comparing large, poorly cut diamonds against small, perfectly cut ones. To see the true benefit of an "Ideal" cut, one would need to compare diamonds of the same carat weight.


```r
ggplot(diamonds, aes(x = carat, y = price, color = cut)) +
  geom_point(alpha = 0.3)
  labs(title = "Price vs. Carat Weight by Cut",
       x = "Carat",
       y = "Price") +
  theme_minimal()
```

# **7.2 Relating the size and carat of a diamond**

## Create a plot to visualize the relationship between the carat and length of a diamond. Zoom in to exclude any outliers.



```r
ggplot(data = diamonds, mapping = aes(x = carat, y = x)) +
  geom_point(alpha = 0.2) + # opacity so we can see clearly the plots
  coord_cartesian(xlim = c(0, 3), ylim = c(3, 10)) + # Zooming in
  labs(
    title = "Relationship between Carat and Diamond Length",
    x = "Carat (Weight)",
    y = "Length (mm)"
  )
```

## What sort of relationship does your plot suggest? How would you explain this relationship?

The plot shows a strong, positive, and slightly curved relationship between carat and length. As a diamond gets heavier (more carats), it naturally gets longer. This happens because weight is tied to volume. Since a diamond is a 3D object, increasing its length usually means it gets wider and deeper too, which causes the weight to jump up quickly. Simply put: more "stuff" inside the diamond makes it both heavier and physically larger.

## Exercise 3.2: Use summarise to determine if there are any diamonds of at least one carat that cost less that $1000.


```r
# Check if any diamonds >= 1 carat cost less than $1000
diamonds %>%
  filter(carat >= 1) %>%
  summarise(any_cheap_large_diamonds = any(price < 1000))
```

## Exercise 4.1: Compute the mean price for diamonds of volume at least one carat.


```r
# Calculate the mean price for diamonds >= 1 carat
mean_price_large_diamonds <- diamonds %>%
  filter(carat >= 1) %>%
  summarise(avg_price = mean(price))

# View the result
mean_price_large_diamonds
```

## Exercise 4.2: Reproduce the output of count(diamonds, cut) via group_by() and summarise().


```r
# Reproducing count(diamonds, cut)
diamonds_count <- diamonds %>%
  group_by(cut) %>%
  summarise(n = n())

# View the result
diamonds_count
```
